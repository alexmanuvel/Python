```python
import os
from sklearn.feature_extraction.text import TfidfVectorizer #Used for word embedding
from sklearn.metrics.pairwise import cosine_similarity #for computing plagiarism
import warnings
from tkinter import *
from tkinter.filedialog import askopenfile
```

```python

class Plagiarism:
    
    file1 = None
    file2 = None
    class_list = []
    output= None
    global s_vector
    def __init__(self, master):
        self.master = master

        ## Creating Frames

        self.top = Frame(self.master, height = 140, bg = 'white')
        self.top.pack(fill = X)

        self.bottom = Frame(self.master, height = 380, bg = '#000000')
        self.bottom.pack(fill = X)

        self.top_image = PhotoImage(file = "check.png", height = 75, width = 60 )
        self.top_image_label = Label(self.top, image = self.top_image, bg = "white")
        self.top_image_label.place(x = 100, y = 30)

        self.heading = Label(self.top, text = "Plagiarism Checker", font = "arial 25 bold", bg = "white", fg = "#000000")
        self.heading.place(x = 200, y = 40)

        self.label = Label(self.bottom, text = "Select",bg = "black", fg = "white", font = "arial 20 bold")
        self.label.place(x = 60, y = 60)

        self.add_button = Button(self.bottom, text = " Open ", bg = "white", fg = "black", font = "arial 15 bold", width = 15, command = self.openFile1, activebackground = "#03fcf8")
        self.add_button.place(x = 70, y = 130)

        self.label2 = Label(self.bottom, text = "Select",bg = "black", fg = "white", font = "arial 20 bold")
        self.label2.place(x = 350, y = 60)

        self.add_button = Button(self.bottom, text = " Open ", bg = "white", fg = "black", font = "arial 15 bold", width = 15, command = self.openFile2, activebackground = "#03fcf8")
        self.add_button.place(x = 370, y = 130)

        self.label3 = Label(self.bottom, text = "Check Your Plagirism Below: ", bg = "black", fg = "white", font = "arial 20 bold")
        self.label3.place(x = 120, y = 230)

        self.add_button = Button(self.bottom, text = " Check Plagiarism", bg = "#24fc03", fg = "black", font = "arial 15 bold", width = 15, command = self.check_plagiarism)
        self.add_button.place(x = 220, y = 300)


    def openFile1(self):

        self.file = askopenfile(mode ='r', filetypes =[('Text File', '*.txt')])
        self.file1 = str(self.file)
        self.file1 = self.file1.split('/')
        self.file2 = self.file1[-1].split("'")
        self.label = Label(self.bottom, text = self.file2[0],bg = "black", fg = "white", font = "arial 20 bold")
        self.label.place(x = 60, y = 60)

        Plagiarism.file1 = self.file

    def openFile2(self):

        self.file = askopenfile(mode ='r', filetypes =[('Text File', '*.txt')])
        self.file1 = str(self.file)
        self.file1 = self.file1.split('/')
        self.file2 = self.file1[-1].split("'")
        self.label2 = Label(self.bottom, text = self.file2[0],bg = "black", fg = "white", font = "arial 20 bold")
        self.label2.place(x = 350, y = 60)

        Plagiarism.file2 = self.file
    def check_plagiarism(self):
        name = Plagiarism.file1.name
        name = name.split('/')
        file1 = name[-1]
        name = Plagiarism.file2.name
        name = name.split('/')
        file2 = name[-1]
        Plagiarism.class_list.append(file1)
        Plagiarism.class_list.append(file2)
        
        warnings.filterwarnings('ignore')
        
        
        files = Plagiarism.class_list
        convert = lambda text: TfidfVectorizer().fit_transform(text).toarray()
        #Convert text to array of numbers
        similarity = lambda doc1, doc2: cosine_similarity([doc1,doc2])
        notes = [open(file).read() for file in files]
        vectors = convert(notes)
        s_vector = list(zip(files, vectors))

        def check():
            result = list()
            
            i = 1
            for x, txt_x in s_vector:
                new_vector = s_vector[i:]
                for y, txt_y in new_vector:
                    sim_score = similarity(txt_x, txt_y)[0][1]
                    names = sorted((x,y))
                    score = (names[0], names[1], sim_score*100)
                    result.insert(0,score)
                i += 1
            Plagiarism.output = result
        check()
        mja_aa_gya = Plagiarism.output
        print(mja_aa_gya) 
        result = Result(mja_aa_gya)
        
        
        
      
    
```

```python
class Result(Toplevel):
    
    def __init__(self, mja_aa_gya):
        Toplevel.__init__(self)

        ## Creating Frames

        self.top = Frame(self, height = 140, bg = 'white')
        self.top.pack(fill = X)

        self.bottom = Frame(self, height = 380, bg = '#000000')
        self.bottom.pack(fill = X)
        
        self.title("Plagiarism Checker")
        self.geometry("650x550+350+150")
        self.resizable(False, False)

        self.top_image = PhotoImage(file = "check.png", height = 75, width = 60 )
        self.top_image_label = Label(self.top, image = self.top_image, bg = "white")
        self.top_image_label.place(x = 100, y = 30)

        self.heading = Label(self.top, text = "Plagiarism Checker", font = "arial 25 bold", bg = "white", fg = "#000000")
        self.heading.place(x = 200, y = 40)
        
        self.label = Label(self.bottom, text = "The Plagiarism Report is: ",bg = "black", fg = "white", font = "arial 20 bold")
        self.label.place(x = 60, y = 60)
        
        statement = mja_aa_gya[0]
        percentage = statement[-1]
        print(percentage)
        
        self.label = Label(self.bottom, text = "{} %".format(percentage),bg = "black", fg = "white", font = "arial 20 bold")
        self.label.place(x =140, y = 130)
        
        
```

```python
def main():
    root = Tk()
    app = Plagiarism(root)
    root.title("Plagiarism Checker")
    root.geometry("650x550+350+150")
    root.resizable(False, False)
    root.mainloop()


if __name__ == '__main__':
    main()
```
